```python
import mdtraj as md
import numpy as np
import matplotlib.pyplot as plt

trajectory = md.load('comp-lab-class-2024/Week2-DataAnalysis/Data/1hz3_T310.stepid25000000.every100ps.nowater.xtc', top='comp-lab-class-2024/Week2-DataAnalysis/Data/1hz3_T310.start.nowater.gro')

atom_indices = [0, trajectory.topology.n_atoms - 1]
end_to_end_distances = md.compute_distances(trajectory, [atom_indices])

radii_of_gyration = md.compute_rg(trajectory)

time = trajectory.time / 1000

plt.figure(figsize=(15, 10))
plt.plot(time, end_to_end_distances, label='End-to-End Distance (nm)', color='blue')
plt.plot(time, radii_of_gyration, label='Radius of Gyration (nm)', color='red')
plt.xlabel('Time (ns)')
plt.ylabel('Distance (nm)')
plt.title('End-to-End Distance and Radius of Gyration Over Time')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
plt.hist(end_to_end_distances, bins=50, density=True, alpha=0.7, color='blue')
plt.title('Normalized Histogram of End-to-End Distances')
plt.xlabel('End-to-End Distance (nm)')
plt.ylabel('Probability Density')

plt.subplot(2, 1, 2)
plt.hist(radii_of_gyration, bins=50, density=True, alpha=0.7, color='red')
plt.title('Normalized Histogram of Radius of Gyration')
plt.xlabel('Radius of Gyration (nm)')
plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()
```

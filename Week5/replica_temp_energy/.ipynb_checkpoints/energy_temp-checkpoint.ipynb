```python
import numpy as np
import matplotlib.pyplot as plt

energy_300 = np.loadtxt('T300_potential_energy.xvg', comments=['@', '#'])[:, 1]
energy_400 = np.loadtxt('T400_potential_energy.xvg', comments=['@', '#'])[:, 1]
energy_600 = np.loadtxt('T600_potential_energy.xvg', comments=['@', '#'])[:, 1]

plt.figure(figsize=(10, 6))
plt.hist(energy_300, bins=30, alpha=0.5, label='T=300K', density=True)
plt.hist(energy_400, bins=30, alpha=0.5, label='T=400K', density=True)
plt.hist(energy_600, bins=30, alpha=0.5, label='T=600K', density=True)

plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Probability Density')
plt.title('Potential Energy Distribution at Different Temperatures')
plt.legend()
plt.show()

data = np.loadtxt('replica_temp.xvg', comments=['@', '#'])
time = data[:, 0]
replica_indices = data[:, 1:]

temperature_map = {0: 300, 1: 400, 2: 600}
temperatures = np.vectorize(temperature_map.get)(replica_indices)

plt.figure(figsize=(10, 6))
for i in range(temperatures.shape[1]):
    plt.plot(time[::100], temperatures[::100, i], label=f'Replica {i+1}')  # 100 

plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Replica Temperature Movement Over Time')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

colvar_data = np.loadtxt('/home/jh9412/HiJihye/comp-lab-class-2024/Week6/Metadynamics/master/Output/ala2_metad_10ns.colvar.txt', skiprows=1)
phi_colvar = colvar_data[:, 1] 
psi_colvar = colvar_data[:, 2]

plt.scatter(phi_colvar, psi_colvar, alpha=0.5, label='COLVAR')
plt.xlabel('Phi (degrees)')
plt.ylabel('Psi (degrees)')
plt.title('Phi vs. Psi')
plt.legend()
plt.savefig('/home/jh9412/HiJihye/comp-lab-class-2024/Week6/Metadynamics/master/Figures/ala1_Phi vs Psi.png')

plt.show()
```

```python
import mdtraj as md

traj = md.load('/home/jh9412/HiJihye/comp-lab-class-2024/week6/input/master/Output/topolB_.xtc', top='/home/jh9412/HiJihye/comp-lab-class-2024/week6/input/master/Output/dialaB.pdb')
phi_indices = md.compute_phi(traj)
psi_indices = md.compute_psi(traj)
phi_mdtraj = md.compute_dihedrals(traj, phi_indices[0])
psi_mdtraj = md.compute_dihedrals(traj, psi_indices[0])

plt.scatter(np.degrees(phi_mdtraj[:, 0]), np.degrees(psi_mdtraj[:, 0]), color='red', alpha=0.5, label='MDTraj')
plt.scatter(phi_colvar, psi_colvar, color='blue', alpha=0.5, label='COLVAR')
plt.xlabel('Phi (degrees)')
plt.ylabel('Psi (degrees)')
plt.title('Phi vs. Psi Comparison')
plt.legend()
plt.savefig('/home/jh9412/HiJihye/comp-lab-class-2024/week6/input/master/Figures/ala1_Phi vs Psi comparison.png')

plt.show()
```

```python
phi_std = np.std(np.degrees(phi_mdtraj[:, 0]))
psi_std = np.std(np.degrees(psi_mdtraj[:, 0]))
print(f"Phi Standard Deviation: {phi_std:.2f}")
print(f"Psi Standard Deviation: {psi_std:.2f}")
```

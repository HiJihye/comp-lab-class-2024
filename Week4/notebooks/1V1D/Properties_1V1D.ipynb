```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the xvg files
end_to_end_data = np.loadtxt('../../Setup/1V1D_AMBER/gromacs/end_to_end_dist.xvg', comments=['#', '@'])
gyration_data = np.loadtxt('../../Setup/1V1D_AMBER/gromacs/gyrate.xvg', comments=['#', '@'])

# Extract time and distance values
time_end_to_end = end_to_end_data[:, 0]
end_to_end_distance = end_to_end_data[:, 1]

time_gyration = gyration_data[:, 0]
radius_of_gyration = gyration_data[:, 1]

# Plotting End-to-End Distance
plt.figure(figsize=(10, 6))
plt.plot(time_end_to_end, end_to_end_distance, label="End-to-End Distance")
plt.xlabel("Time (ps)")
plt.ylabel("Distance (nm)")
plt.title("End-to-End Distance vs Time")
plt.grid(True)

plt.savefig('../../Figures/1V1D/End-to-End Distance_Time.png')
plt.show()

# Plotting Radius of Gyration
plt.figure(figsize=(10, 6))
plt.plot(time_gyration, radius_of_gyration, label="Radius of Gyration", color="green")
plt.xlabel("Time (ps)")
plt.ylabel("Radius of Gyration (nm)")
plt.title("Radius of Gyration vs Time")
plt.grid(True)

plt.savefig('../../Figures/1V1D/Gyration_Time.png')
plt.show()
```
